```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("homeprices.csv")
df.head()
```

```python
dummy = pd.get_dummies(df.town)
dummy

#here we are doing the one hot encoding and giving the value to the nominal data
```

```python
merge = pd.concat([df,dummy],axis = 'columns')
merge

#Here we are merging the original table to the encoded table;
```

```python
final = merge.drop(['town','west windsor'],axis = 'columns')
final

#Here we are removing\Droping the unwanted columns
```

```python
from sklearn.linear_model  import LinearRegression
r = LinearRegression()
```

```python
#provide the X and Y for the prediction;

X = final.drop(['price'],axis = 'columns')
X
```

```python
# Y = price because the price is the independent variable it will depend on the other factors

Y = final.price
Y
```

```python
# Train the model

r.fit(X,Y)
```

```python
r.predict([[1,0,2800]])
```
