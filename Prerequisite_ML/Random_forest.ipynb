```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib  inline
from sklearn.datasets import load_digits
```

```python
dig = load_digits()
dir(dig)
```

```python
dig.data
```

```python
plt.gray()
for i in range (5):
    plt.matshow(dig.images[i])
```

```python
dig.target
```

```python
df = pd.DataFrame(dig.data)
df.head()
```

```python
df['target'] = dig.target
df.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(df.drop(['target'],axis = 'columns'),df.target,test_size = 0.2,random_state = 10)
```

```python
from sklearn.ensemble import RandomForestClassifier
obj = RandomForestClassifier(n_estimators = 50)
```

```python
obj.fit(x_train,y_train)
```

```python
obj.score(x_test,y_test)
```

```python
y_predict = obj.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test,y_predict)
cm
```

```python
import seaborn as sb
plt.figure(figsize=(10,7))
sb.heatmap(cm,annot = True)         #annot=True: Annotates each cell with its value, which is useful for precise data interpretation.
plt.xlabel("predicted")
plt.ylabel("Truth")
```
