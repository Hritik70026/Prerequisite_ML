```python
import pandas as pd
import numpy as np
from sklearn.datasets import load_iris
iris = load_iris()
```

```python
dir(iris)
```

```python
iris.feature_names
```

```python
##  Here we are making the table for the flower where feature are the columns and the rows are the data  ##
 
df = pd.DataFrame(iris.data,columns=iris.feature_names)
df.head()
```

```python
## here we are adding the new column in our table ##

df['target'] = iris.target
df.head()
```

```python
iris.target_names
```

```python
## check the target == 1

df[df.target == 1].head()
```

```python
## to add the new column "FLOWER NAME " from one column "TARGET" to generate another column we use the "Apply" function in pandas and in apply functtion we use "lambda" function ##

df['Flower_names'] = df.target.apply(lambda x : iris.target_names[x])
df.head()
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df0 = df[df.target == 0]
df1 = df[df.target == 1]
df2 = df[df.target == 2]
```

# Ploting the graphs for visualization

```python
# plt.xlabel(['sepal length (cm)'])
# plt.ylabel('sepal width (cm)')
plt.scatter(df0['sepal length (cm)'],df0['sepal width (cm)'],color = 'green',marker = '+',s = 25)
plt.scatter(df1['sepal length (cm)'],df1['sepal width (cm)'],color = 'red',marker = '*',s = 25)
```

```python

plt.scatter(df0['petal length (cm)'],df0['petal width (cm)'],color = 'blue',marker = '*')
plt.scatter(df1['petal length (cm)'],df1['petal width (cm)'],color = 'green',marker = '+')
```

# Trianing the model

```python
from sklearn.model_selection import train_test_split
```

```python
x = df.drop(['target','Flower_names'],axis = 'columns')
x.head()
```

```python
y = (df.target)
y.head()
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=10)
```

```python
from sklearn.svm import SVC
obj = SVC()
```

```python
## Trainng of the model

obj.fit(x_train,y_train)
```

```python
obj.score(x_train,y_train)
```

```python
obj.predict([[10.5,8.6,1.2,4.5]])
```

# Tuning of the parameters


```python
obj_reg = SVC(C=1)
obj_reg.fit(x_train,y_train)
obj_reg.score(x_test,y_test)
```
